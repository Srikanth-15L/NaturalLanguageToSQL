````python
import os
import requests
from langchain_openai import ChatOpenAI
from langchain_core.tools import tool
````

# 1. Load API Keys

Create a `.env` file in the project root with:
```
OPENAI_API_KEY=sk-...
WEATHER_API_KEY=your_weatherapi_key
```
Get a free weather API key at https://www.weatherapi.com

````python
import os
from dotenv import load_dotenv

load_dotenv()

assert os.getenv("OPENAI_API_KEY"), "Set OPENAI_API_KEY in your .env file"
WEATHER_API_KEY = os.getenv("WEATHER_API_KEY")
assert WEATHER_API_KEY, "Set WEATHER_API_KEY in your .env file"

print("Keys loaded.")
````

# 2. Create 3rd Party Tool

````python
@tool
def get_current_weather(location:str)->dict:
    """This tool gives you the current weather at any location in the world """
    url="http://api.weatherapi.com/v1/current.json"
    params={"key":WEATHER_API_KEY,"q":location}
    response=requests.get(url,params=params)
    return response.json()
````

# 3. Tool Binding

````python
llm=ChatOpenAI(model="gpt-4",temperature=0)
mode_with_tool=llm.bind_tools([get_current_weather])
````

````python
# Ask a Question

user_input="What is the Temperature in Dallas?"
response=mode_with_tool.invoke(user_input)
print(response)
````

# 4. Extract Input from JSON and Pass to tool

````python
if hasattr(response,"tool_calls") and response.tool_calls:
    print("Tool Call Made by LLM:")
    print(response.tool_calls)

    
````

````python
for call in response.tool_calls:
    tool_name=call["name"]
    args=call["args"]

    if tool_name=="get_current_weather":
        tool_result=get_current_weather.invoke(args)
        print("Tool Result:")
        print(tool_result)
````

````python
from langchain.prompts import PromptTemplate

weather_prompt="""
You are a helpful weather assistant. Answer the user question in a nice professional format

User Question:{question}

Answer: {response}

Convert to a Professional answer
"""

prompt=PromptTemplate(
    input_variables=["question","response"],
    template=weather_prompt)

chain=prompt | llm

answer=chain.invoke({"question":user_input ,"response":tool_result} )

print(answer.content)
````
